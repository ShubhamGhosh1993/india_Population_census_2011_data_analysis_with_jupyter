```python
import pandas as pd
import os as os
# Read the data and make dataframe.
df = pd.read_csv(f"{os.getcwd()}/data/india-districts-census-2011.csv")
```

```python
# Set the district code as indexes
df.set_index("District code")
df.style.set_caption("Indian Census 2011")
df
```

```python
#Show data related to Districts Mumbai,New Delhi and Kolkata.
df[df["District name"].isin(["Mumbai","New Delhi","Kolkata"])]
```

```python
# Calculate state wise 
# Total populations 
stateWise_TotalPopulations = df.groupby('State name')['Population'].sum().sort_values(ascending=True)
print(stateWise_TotalPopulations)
```

```python
### Calculate state wise
# Religion wise total populations
df.groupby('State name')['Hindus', 'Muslims','Christians','Jains', 'Buddhists','Sikhs', 'Others_Religions','Religion_Not_Stated'
].sum().sort_values(by='State name',ascending=True)
```
